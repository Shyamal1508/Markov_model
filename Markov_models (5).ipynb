```python
sherlock_path = 'C:/Users/bhatt/OneDrive/Desktop/whisper-main/sherlock/'

def read_all_stories(path):
    
    txt = []
    for _, _, files in os.walk(path):
        for file in files:
            with open(path + file) as f:
                for line in f:
                    line = line.strip()
                    if line == '----------':
                        break
                    if line != '':
                        txt.append(line)
    return txt
```

```python
all_lines = read_all_stories(sherlock_path)
```

```python
def clean_txt(lines):
    cleaned_txt = []
    for line in lines:
        line = line.lower()
        line = re.sub(r"[,\'\"/\\@!.+=_?()--|-]", '', line)
        tokens = word_tokenize(line)
        words = [word for word in tokens]
        cleaned_txt+=words
    return cleaned_txt
```

```python
cleaned_lines = clean_txt(all_lines)
```

```python
n_gram = 2

markov_model = {}

for i in range(100):
    curr_state, next_state = '', ''
    for j in range(n_gram):
        curr_state += cleaned_lines[i+j] + " "
        next_state += cleaned_lines[i+j+n_gram] + " "
    curr_state = curr_state[:-1]
    next_state = next_state[:-1]
    
    if curr_state not in markov_model:
        markov_model[curr_state] = {}
        markov_model[curr_state][next_state] = 1
    else:
        if next_state in markov_model[curr_state]:
            markov_model[curr_state][next_state] += 1
        else:
            markov_model[curr_state][next_state] = 1
    print(markov_model)
```

```python
for curr_state, transition_state in markov_model.items():
    total_sum = sum(transition_state.values())
    for state, count in transition_state.items():
        markov_model[curr_state][state] = count/total_sum
```
